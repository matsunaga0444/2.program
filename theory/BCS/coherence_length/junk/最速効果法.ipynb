```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
 
# 
def f(x, y):
    z = np.sin(x) + np.cos(y)
    return z
# 
def df(x, y):
    dzdx = np.cos(x)
    dzdy = - np.sin(y)
    dz = np.array([dzdx, dzdy])
    return dz
 
# 
eta = 0.4                           # 
max_iteration = 1000                # 
x0 = 2                              # x0
y0 = -0.1                           # y0
x_pred = [x0]                       # x0()
y_pred = [y0]                       # y0()
 
# 
for i in range(max_iteration):
    x0, y0 = np.array([x0, y0]) - eta * df(x0, y0)          # 
    x_pred.append(x0)               # x0
    y_pred.append(y0)               # y0
    print(i, x0, y0)
 
x_pred = np.array(x_pred)           # x0numpy
y_pred = np.array(y_pred)           # x0numpy
z_pred = f(x_pred, y_pred)          # z
 
# 
x = np.arange(0, 6, 0.2)
y = np.arange(-5, 1, 0.2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
 
# ----------------------------------------------------------------
# 
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'
 
#  
fig = plt.figure()
ax1 = Axes3D(fig)
 
# 
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
 
# 
ax1.plot_wireframe(X, Y, Z, label='f(x, y)')
ax1.scatter3D(x_pred, y_pred, z_pred, label='gd', color='red', s=100)
plt.savefig(".png")

# 
plt.show()

```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
 
# 
def f(x, y):
    z = x ** 2 + y ** 2
    return z
# 
def df(x, y):
    dzdx = 2 * x
    dzdy = 2 * y
    dz = np.array([dzdx, dzdy])
    return dz
 
# 
eta = 0.1                           # 
max_iteration = 1000                # 
x0 = -10                            # x0
y0 = 10                             # y0
x_pred = [x0]                       # x0()
y_pred = [y0]                       # y0()
 
# 
for i in range(max_iteration):
    x0, y0 = np.array([x0, y0]) - eta * df(x0, y0)          # 
    x_pred.append(x0)               # x0
    y_pred.append(y0)               # y0
    print(i, x0, y0)
 
x_pred = np.array(x_pred)           # x0numpy
y_pred = np.array(y_pred)           # x0numpy
z_pred = f(x_pred, y_pred)          # z
 
# 
x = np.arange(-10, 11, 2)
y = np.arange(-10, 11, 2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
 
# ----------------------------------------------------------------
# 
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'
 
#  
fig = plt.figure()
ax1 = Axes3D(fig)
 
# 
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
 
# 
ax1.plot_wireframe(X, Y, Z, label='f(x, y)')
ax1.scatter3D(x_pred, y_pred, z_pred, label='gd', color='red', s=50)
 
# 
plt.show()
plt.close()
# ---------------------------------------------------------------------------------
```
