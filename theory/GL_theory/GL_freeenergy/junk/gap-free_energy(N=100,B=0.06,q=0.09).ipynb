```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#パラメーターの代入
a, b, h, m, Tc ,T = 1, 1, 1, 1, 0.08, 0.12

## GL_free_energyをdef
def GL_F(gap,q):
    return a*(T-Tc)*gap**2 + (b/2)*gap**4 + (h**2/2*m)*gap**2*q**2


#GL_free_energyの計算と描画-----------------------------------------------------------------------------------
# Figureと3DAxeS
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection="3d")

# 軸ラベルを設定
ax.set_xlabel("gap", size = 16)
ax.set_ylabel("q", size = 16)
ax.set_zlabel("free_energy", size = 16)

# (x,y)データを作成
gaps = np.linspace(-0.075,0.075,50) 
qs   = np.linspace(-0.075,0.075,50) 

# 格子点を作成
GAPs, Qs = np.meshgrid(gaps, qs)

# 高度の計算式
Fs = GL_F(GAPs, Qs)

# 曲面を描画
ax.plot_surface(GAPs, Qs, Fs, cmap = "summer")

# 底面に等高線を描画
ax.contour(GAPs, Qs, Fs, colors = "black", offset = 0)

plt.show()
```
